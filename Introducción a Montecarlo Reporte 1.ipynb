**1. Calculadora de integrales mediante Monte Carlo**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_a_integrar(x):
    return 8*x**4

a = 10
b = 15
N = 100

x = np.random.uniform(a, b, N)
y = np.random.uniform(0, funcion_a_integrar(b), N)

n_bajo_curva = np.sum(y < funcion_a_integrar(x))
respuesta = (b - a) * funcion_a_integrar(b) * n_bajo_curva / N
respuesta
```

```python
plt.plot(x, y, '.')
plt.plot(np.linspace(a, b, 100), funcion_a_integrar(np.linspace(a, b, 100)), 'r')
plt.show()
```

**2. Estimación de π usando el método del círculo**

```python
N = 100000 
x = np.random.uniform(-1, 1, N)
y = np.random.uniform(-1, 1, N)
dentro = x**2 + y**2 <= 1
pi = 4 * np.sum(dentro) / N
pi
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x[dentro], y[dentro], s=1)
plt.scatter(x[~dentro], y[~dentro], s=1)
plt.gca().set_aspect('equal')
plt.legend(markerscale=5)
plt.show()
```
