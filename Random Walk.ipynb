```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n_steps =200
steps = np.random.choice([-1,1], size=n_steps)
price =100 + np.cumsum(steps)

plt.figure(figsize=(10,5))
plt.plot(price)
plt.title("Random Walk de un precio")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

n_sim =5000
n_steps =200

final_positions = []

for _ in range(n_sim):
    steps = np.random.choice([-1,1], size=n_steps)
    final_positions.append(np.sum(steps))

plt.hist(final_positions, bins=50)
plt.title("Distribucin de posiciones finales")
plt.show()
```

```python
import numpy as np

np.random.seed(0)

n_sim = 2000
times = [10, 50, 100, 200, 400]

for t in times:
    finals = []
    for _ in range(n_sim):
        steps = np.random.choice([-1, 1], size=t)
        finals.append(np.sum(steps))
    print(f"t={t}, var={np.var(finals):.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

T =1
n =1000
dt = T/n

increments = np.random.normal(0, np.sqrt(dt), size=n)
W = np.cumsum(increments)

plt.plot(W)
plt.title("Aproximacin a movimiento browniano")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n_walks = 200
n_steps = 200
x0 = 100 

steps = np.random.choice([-1, 1], size=(n_walks, n_steps))
walks = x0 + np.cumsum(steps, axis=1)

mean_walk = walks.mean(axis=0)

plt.figure(figsize=(10, 5))
plt.plot(walks.T, alpha=0.25)  
plt.title("200 caminatas aleatorias (pasos 1)")
plt.xlabel("Tiempo (t)")
plt.ylabel("Posicin / Precio")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(mean_walk, linewidth=2)
plt.title("Promedio de 200 caminatas en cada tiempo t")
plt.xlabel("Tiempo (t)")
plt.ylabel("Promedio")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(walks.T, alpha=0.15)
plt.plot(mean_walk, linewidth=3)  # promedio resaltado
plt.title("200 caminatas + su promedio")
plt.xlabel("Tiempo (t)")
plt.ylabel("Posicin / Precio")
plt.show()
```
